```python
from scipy.stats import  chi2_contingency
t=pd.crosstab(df['rating_count'],df['rating'])
chi2,p_value,dof,expected=chi2_contingency(t)
print('chi_square :',chi2)
print('P_value :',p_value)
print('degree of freedom :',dof)
print('Expected frequancies:',expected)
```

```python
from scipy.stats import  f_oneway
df=df.dropna(subset=['rating','category'])
category=df['category'].unique()
group=[df[df['category']==cat]['rating']for cat in category]
f_stat,p_value=f_oneway(*group)
print('F_statistic:',f_stat)
print('p_value:',p_value)
```
